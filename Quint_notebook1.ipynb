```python
df_claims = pd.read_excel('Claim_information_per_state_final.xlsx', sheet_name='Main Table',  engine='openpyxl')
```

```python
# Calculate the averages for each state
avg_insured = df_claims.groupby('State')['Insured drivers (%)'].mean()
avg_licensed = df_claims.groupby('State')['Licensed drivers'].mean()

# Calculate the product for each state and map it to the original dataframe
df_claims['Insured drivers'] = df_claims['State'].map(avg_insured * avg_licensed / 100)
```

```python
df_claims.to_excel('df_claims.xlsx', index=False, engine='openpyxl')
```

```python
df_claims_melted = pd.melt(df_claims, id_vars=['State','Year'])
print(df_claims_melted)
```

```python
df_claims_melted.to_excel('df_claims_melted.xlsx', index=False, engine='openpyxl')
```

```python
df_claims['Average_tlpr'] = 100 - df_claims.groupby('State')['Total liablity loss ratio'].transform('mean')
```

```python
df_claims['Average_blpr'] = 100 - df_claims.groupby('State')['Boduly injury loss ratio'].transform('mean')
```

```python
df_claims['Gross profit per driver'] = (df_claims['Total liability premiums']- df_claims['Total liablity inc losses']) / df_claims['Insured drivers'] 
df_claims['Average_gppd'] = df_claims.groupby('State')['Gross profit per driver'].transform('mean')
```

```python
df_claims['Bodily claims per inc loss'] = df_claims['Bodily injury inc claims']/df_claims['Bodily injury inc losses']
df_claims['Average_cpil'] = df_claims.groupby('State')['Bodily claims per inc loss'].transform('mean')
```

```python
df_claims = df_claims.reset_index(drop=True)
```

```python
# Function to assign ratings based on the difference between min and max values
def assign_rating(column):
    min_value = column.min()
    max_value = column.max()
    
    # Calculate the range between min and max divided by 5
    value_range = (max_value - min_value) / 5
    
    # Assign a rating based on the value's position within the range
    def inner_assign_rating(value):
        if value <= min_value + value_range:
            return 1
        elif value <= min_value + 2 * value_range:
            return 2
        elif value <= min_value + 3 * value_range:
            return 3
        elif value <= min_value + 4 * value_range:
            return 4
        else:
            return 5
    
    return inner_assign_rating
```

```python
# Create a new "Rating" column by applying the function
df_claims['Rating_Total_Liability_Profit_Ratio'] = df_claims['Average_tllr'].apply(assign_rating(df_claims['Average_tllr']))
df_claims['Rating_Bodily_Liability_Profit_Ratio'] = df_claims['Average_blpr'].apply(assign_rating(df_claims['Average_blpr']))
df_claims['Rating_Gross_Profit_Per_Driver'] = df_claims['Average_gppd'].apply(assign_rating(df_claims['Average_gppd']))
df_claims['Rating_Bodily_Cost_Per_Claim'] = df_claims['Average_cpil'].apply(assign_rating(df_claims['Average_cpil']))


df_claims.to_excel('df_claims_ratings.xlsx', index=False, engine='openpyxl')
```
